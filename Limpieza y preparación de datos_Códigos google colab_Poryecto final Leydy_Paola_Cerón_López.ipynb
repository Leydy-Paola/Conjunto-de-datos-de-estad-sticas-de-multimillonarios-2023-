##Asegurandose que las edades sean números enteros (sin decimales)

Pero antes vamos a crear una manera más simplificada de visualizar los datos de nuestro dataset con df.head

```python
#Este código imprimirá las primeras 5 filas del DataFrame df.
df.head()
```

Como pudimos visualizar los datos de la columna **"age"** están en decimales, por lo que realizaremos el cambio a números enteros, como lo mostraremos a continuación;

```python
# Convierte la columna 'age' a tipo de dato entero
df['age'] = df['age'].astype(int)
df.head()
```

Eliminamos columnas filtradas con datos nulos y que no pertenecen a nuestro conjunto de datos requeridos

```python
# Lista de nombres de las columnas que deseas eliminar
columnas_a_eliminar2 = ['date']

# Elimina las columnas especificadas del DataFrame
df = df.drop(columns=columnas_a_eliminar2)
```

```python
# Lista de nombres de las columnas que deseas eliminar
columnas_a_eliminar3 = ['birthYear','birthMonth','birthDay']

# Elimina las columnas especificadas del DataFrame
df = df.drop(columns=columnas_a_eliminar3)
```

##Cambiando formato a los datos



Ahora pasamos a eliminar la parte de la hora (0:00) de la columna 'birthDate', para esto usamos el método pd.to_datetime() para convertir la columna a objetos de tipo datetime y luego usar dt.date para quedarnos solo con la parte de la fecha y eliminar la parte de la hora

```python
# Convertimos la columna 'birthDate	' a formato datetime y luego a solo fecha y visualizamos nuevamente
df['birthDate'] = pd.to_datetime(df['birthDate']).dt.date
df.head()
```

```python
#Visualizamos los tipos de datos de nuestra dataset
print(df.dtypes)
```

Es preferible manejar un solo tipo de datos, sin embargo el cambio de algunos datos en sus formatos, se irán ajustando en el momento de visualizarlo con PowerBi

```python
# Convierte la columna 'birthDate' a objetos de fecha y hora
df['birthDate'] = pd.to_datetime(df['birthDate'])

# Verifica que la conversión se haya realizado correctamente
print(df['birthDate'])
```

##Ordenando en orden ascendente

```python
# Ordena el DataFrame por la columna 'rank' en orden ascendente
df = df.sort_values(by='rank', ascending=True)

# Muestra las primeras filas del DataFrame ordenado
df.head()
```

##Convertir el número en tu conjunto de datos a dólares con dos ceros demás antes de la coma

```python
# Multiplica el número por 100 para agregar dos ceros antes de la coma
df['finalWorth'] = df['finalWorth'] * 100

# Convierte el número a dólares y formatea con dos ceros antes de la coma
df['finalWorth'] = df['finalWorth'].apply(lambda x: '${:,.2f}'.format(x))

# Imprime el DataFrame para verificar los cambios
print(df)
```

```python
#Visualizamos de manera más clara el cambio obtenido
df.head()
```

##Cambiando nombres de columnas

```python
# Definimos un diccionario que mapee los nombres antiguos a los nuevos nombres
nuevos_nombres = {
    "residenceStateRegion": "Región de residencia",
    "gender": "Género",
    "countryOfCitizenship": "País de ciudadanía ",
    "personName": "Nombre",
    "finalWorth": "Patrimonio neto",
    "source": "Fuente de riqueza",
}

# Cambiamos los nombres de las columnas usando el método rename()
df.rename(columns=nuevos_nombres, inplace=True)

# Mostramos las primeras filas para verificar los cambios
df.head()
```
